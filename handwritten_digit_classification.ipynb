```
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```
(x_train , y_train) , (x_test , y_test) = keras.datasets.mnist.load_data()
```

```
x_train = x_train / 255
x_test = x_test / 255
```

```
len(x_test)
```

```
plt.matshow(x_train[5])
```

```
y_train[5]
```

```
x_train.shape
```

```
#converting the matrix into single dimension for input layer
x_train_reshape = x_train.reshape(len(x_train) , 28*28)
x_test_reshape = x_test.reshape(len(x_test) , 28*28)
```

```
x_train_reshape.shape
```

```
model = keras.Sequential([
    keras.layers.Dense(10 , input_shape = (784 ,) , activation = 'sigmoid')
])

model.compile(optimizer = 'adam' ,
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy']
              )

# for tranning part
model.fit(x_train_reshape , y_train , epochs = 10)
```

```
#test accuracy on test dataset
model.evaluate(x_test_reshape , y_test)
```

```
plt.matshow(x_test[299])
```

```
y_predicted = model.predict(x_test_reshape)
y_predicted[299]
```

```
y_predicted_labels = [np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```
y_test[:5]
```

```
np.argmax(y_predicted[299])
```

```
cm = tf.math.confusion_matrix(labels=y_test , predictions = y_predicted_labels)
```

```
import seaborn as sns
plt.figure(figsize=(10 , 7))
sns.heatmap(cm , annot = True , fmt = 'd')
plt.xlabel('predicted')
plt.ylabel('Truth')
```

```
#adding hidden layers in the above model
model = keras.Sequential([
    keras.layers.Dense(100 , input_shape = (784 ,) , activation = 'relu'),
    keras.layers.Dense(10 , input_shape = (784 ,) , activation = 'sigmoid')
])

model.compile(optimizer = 'adam' ,
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy']
              )

# for tranning part
model.fit(x_train_reshape , y_train , epochs = 10)
```

```
model.evaluate(x_test_reshape , y_test)
```
